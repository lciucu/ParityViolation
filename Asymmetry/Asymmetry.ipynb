```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
debug=False

step_theta=0.01
min_theta=0.0
max_theta=np.pi

step_beta=0.01
min_beta=0.0
max_beta=1.0

outputFolderName="../Report"
```

```python
def asymmetry(b2,b4,c2,c4):
    A=(1-c2+b2*(2-2*c4)+b4*(4-5*c2+c4))/(1+3*c2+b2*(2+6*c2)+b4*(6-3*c2+c4))
    if debug:
        print("A",A,"b2",b2,"b4",b4,"c2",c2,"c4",c4)
    return A
# done function
```

```python
list_theta=[]
list_beta=[]
list_A=[]
theta=min_theta
while theta<=max_theta:
    theta+=step_theta
    cos_theta=np.cos(theta)
    cos_theta_2=cos_theta*cos_theta
    cos_theta_4=cos_theta_2*cos_theta_2
    #print(theta,cos_theta,cos_theta_2,cos_theta_4)
    beta=min_beta
    while beta<=max_beta:
        beta+=step_beta
        beta_2=beta*beta
        beta_4=beta_2*beta_2
        #print(beta,beta_2,beta_4)
        A=asymmetry(beta_2,beta_4,cos_theta_2,cos_theta_4)
        list_theta.append(theta)
        list_beta.append(beta)
        list_A.append(A)
    # done while loop over beta
# done while loop over theta 
nparray_theta=np.array(list_theta)
nparray_beta=np.array(list_beta)
nparray_A=np.array(list_A)    
```

```python
#nparray_theta_2=nparray_theta[::10]
#nparray_beta_2=nparray_beta[::10]
#nparray_A_2=nparray_A[::10]
#
nparray_theta_2=nparray_theta
nparray_beta_2=nparray_beta
nparray_A_2=nparray_A
#
fig=plt.figure(figsize=(12,6))
ax=fig.add_subplot(1,1,1,projection='3d')
ax.bar3d(nparray_theta_2,nparray_beta_2,np.zeros(len(nparray_A_2)),step_theta*10,step_beta*10,nparray_A_2,shade=True)
ax.set_title("Asymmetry vs Theta and Beta")
ax.set_xlabel("Theta angle, from 0 to Pi")
ax.set_ylabel("Beta factor, from 0 to 1")
ax.set_zlabel("Asymmetry")
plt.savefig(outputFolderName+"/Asymmetry_vs_Theta_vs_Beta.png")
```

```python
list_fixedtheta_beta=[]
list_fixedtheta_A=[]
theta=np.pi/2
cos_theta=np.cos(theta)
cos_theta_2=cos_theta*cos_theta
cos_theta_4=cos_theta_2*cos_theta_2
beta=min_beta
while beta<=max_beta:
    beta+=step_beta
    beta_2=beta*beta
    beta_4=beta_2*beta_2
    #print(beta,beta_2,beta_4)
    A=asymmetry(beta_2,beta_4,cos_theta_2,cos_theta_4)
    list_fixedtheta_beta.append(beta)
    list_fixedtheta_A.append(A)
# done while loop over beta
nparray_fixedtheta_beta=np.array(list_fixedtheta_beta)
nparray_fixedtheta_A=np.array(list_fixedtheta_A)    
```

```python
plt.plot(nparray_fixedtheta_beta,nparray_fixedtheta_A)
plt.xlabel("Beta factor from 0 to 1")
plt.ylabel("Asymmetry")
plt.title("Fixed Theta=Pi/2")
plt.savefig(outputFolderName+"/Asymmetry_vs_Beta.png")
```
